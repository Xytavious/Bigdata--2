```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime 
```

```python
df =pd.read_csv('all_stocks_5yr.csv')
df = df.dropna()
sorted_df = df.groupby('date')['close'].mean().plot()
plt.title("All stocks from 2013-2017")
plt.xlabel("Date")
plt.ylabel("closing price")
plt.xticks(rotation=45)
sorted_df
```

Here are the avagrage closing prices for all stocks from 2013-2017
From this data you can tell that form 2013-2017 avrage cloding price per stock has increased

```python
sd = pd.read_csv('all_stocks_5yr.csv')
sd=sd.dropna()

avg_cl_per = sd.groupby('Name')['close'].mean()
avg_cl_per = avg_cl_per.sort_values(ascending= False)

plt.Figure(figsize=(12,16))


top = avg_cl_per.head(10)
bottom = avg_cl_per.tail(10)

comp = pd.DataFrame({
    'Top': top,
    'Bottom': bottom})
comp.plot(kind='bar')
plt.title('average closing per stock 5yr')
plt.xlabel('Compaines')
plt.ylabel('Avarage closing price')
plt.xticks(rotation=45)
plt.tight_layout
plt.show
```

Here is the avaerage closing prices for the Top 10 stocks vs the Bottom 10 stocks 
As shown from the chart the top 10 stocks have much higher closing prices than the bottom 10

```python
v = pd.read_csv('all_stocks_5yr.csv')
v =v.dropna()
stock_stats= v.groupby('Name').agg({
    'close': 'mean',
    'volume': 'mean'
}).rename(columns={'close': 'avg_close', 'volume': 'avg_volume'})

top2 = stock_stats.sort_values(by='avg_close', ascending=False).head(100)

plt.figure(figsize=(5,10))
plt.scatter(top2['avg_close'], top2['avg_close'])

for s in top2.index:
    plt.annotate(s, (top2.loc[s,'avg_close'],top2.loc[s, 'avg_volume']))

plt.title('top 100 stocks(Avg close / Acg Value)')
plt.xlabel('Closing Price')
plt.ylabel('Avagre Value')
plt.grid(True)
plt.show()
```

This shows the top 100 stocks avarage closing prices 
This shows an steady up trend based on the stock value
