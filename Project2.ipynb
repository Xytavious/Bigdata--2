```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime 
```

```python
df =pd.read_csv('all_stocks_5yr.csv')
df = df.dropna()
sorted_df = df.groupby('date')['close'].mean().plot()
plt.title("All stocks from 2013-1017")
plt.xlabel("Date")
plt.ylabel("closing price")
plt.xticks(rotation=45)
sorted_df
```

```python
sd = pd.read_csv('all_stocks_5yr.csv')
sd=sd.dropna()

avg_cl_per = sd.groupby('Name')['close'].mean()
avg_cl_per = avg_cl_per.sort_values(ascending= False)

plt.Figure(figsize=(12,16))


top = avg_cl_per.head(10)
top.plot(kind='bar')
plt.title('average closing per stock 5yr')
plt.xlabel('Compaines')
plt.ylabel('Avarage closing price')
plt.xticks(rotation=45)
plt.tight_layout
plt.show
```

```python
v = pd.read_csv('all_stocks_5yr.csv')
v =v.dropna()
stock_stats= v.groupby('Name').agg({
    'close': 'mean',
    'volume': 'mean'
}).rename(columns={'close': 'avg_close', 'volume': 'avg_volume'})

top2 = stock_stats.sort_values(by='avg_close', ascending=False).head(100)

plt.figure(figsize=(5,10))
plt.scatter(top2['avg_close'], top2['avg_close'])

for s in top2.index:
    plt.annotate(s, (top2.loc[s,'avg_close'],top2.loc[s, 'avg_volume']))

plt.title('top 100 stocks(Avg close / Acg Volume)')
plt.xlabel('Closing Price')
plt.ylabel('Avagre Volume')
plt.grid(True)
plt.show()
```
